```python
from scipy.io import wavfile
import numpy as np

sample_rate,data = wavfile.read('sound wave.wav')

print(f"Sample rate : {sample_rate} Hz")
print(f"Data type : {data.dtype}")
print(f"Data shape : {data.shape}")
```

```python
signal = data
L = len(data)
T = 1/sample_rate
t = np.arange(L) * T
print('t shape:',t.shape)
print("sample of t\n",t[1:10])
```

```python
t_plot = 5
msec_per_sce = 1000
numsamples = t_plot * sample_rate / msec_per_sce
print("numsamples:",numsamples)
print("numsamples type:",type(numsamples))
t_msec = t*msec_per_sce
```

```python
numsamples = int(numsamples)
print('numsamples:',numsamples)
print('numsamples type:',type(numsamples))
```

```python
import matplotlib.pyplot as plt
plt.figure(1)
%matplotlib inline
plt.plot(t_msec[:numsamples],signal[:numsamples])
plt.title("Sample of the signal")
plt.xlabel('time(millisecond)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
yf = np.fft.fft(signal)
xf = np.fft.fftfreq(L,1/sample_rate)
half_N = L//2
xf_half = xf[:half_N]
yf_half = yf[:half_N]
magnitude_spectrum = np.abs(yf_half)
plt.figure(2)
plt.plot(xf_half,magnitude_spectrum)
plt.title('Magnitude Spectrum')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.tight_layout()
plt.show()
```
